```python
#import libraries
import pandas as pd
import re

#load the dataset
df1 = pd.read_csv('datasets/logins.csv')
df1.head(10)
```

```python
len(df1)
```

```python
#splitting the username column into First and Last name for easy analysis
Names = df1.username.str.split('.', n = 1, expand = True)

#create new columns for the first and last name
df1['FirstName'] = Names[0]
df1['LastName'] = Names[1]
```

## Condition 1: Name must not be present in password

```python
#compare the name columns with the password columns for name presence and keep rows without names in password
new_df = []
for index, row in df1.iterrows():
    if row['FirstName'] not in row['password'].lower() and row['LastName'] not in row['password'].lower():
        new_df.append(row['password'])
```

```python
len(new_df)
```

```python
#filter out rows that do not have one of the usernames in the password
df = df1[df1['password'].isin(new_df)]
```

## Condition 2: Special characters that must be present in the password



```python
#The password column must meet the following conditions
# 1. len > 10
# 2. must contain numbers
# 3. must contain non-alphanumeric values
# 4. must not contain the phrase 'password'
# 5. must contain small and capital letters

result = df[df.password.str.contains('\W') & df['password'].apply(lambda x: len(x) >= 10) & ~df.password.str.contains('password', flags=re.IGNORECASE) & df.password.str.contains('[A-Z]') & df.password.str.contains('[a-z]') & df.password.str.contains('[0-9]')]
```

## Finding the percentage of users with invalid passwords

```python
bad_pass = round((len(df1) - len(result)) / len(df1), 2)  

print('The percentage of users that have invalid passwords is {} '.format(bad_pass))
```
